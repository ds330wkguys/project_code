```python
# import dependencies
import pandas as pd
import snscrape.modules.twitter as sntwitter
import itertools
from nltk import *
import re
import string                             
```

```python
# search Apple stock data in twitter that generated within one year.
search_word = "AAPL"

start_day = "2021-10-01"
end_day = "2022-10-01"

search_query = search_word + ' since:' + start_day + ' until:' + end_day 
```

```python
# crawling 1000 data
scraped_tweets = sntwitter.TwitterSearchScraper(search_query).get_items()
sliced_scraped_tweets = itertools.islice(scraped_tweets, 1000)
```

```python
# organizing the data as dataframe
df = pd.DataFrame(sliced_scraped_tweets)
df
```

```python
# creating test dataframe
df_tmp = df[['date','renderedContent']].copy()
```

```python
df_tmp
```

```python
import preprocessor as p
```

```python
# adding each rows in a list
list_content = []
for i in df_tmp['renderedContent']:
    list_content.append(i)
```

```python
list_content
```

```python
# in every element of the list, process the data cleaning
tmp = []
for i in list_content:
    tmp.append(p.clean(i))
```

```python
# removing all other languages except english 
result = []
for i in tmp:
    text = re.sub('[^a-zA-Z0-9]',' ',i).strip()
    if(text != ''):
        result.append(text)
```

```python
df_test = pd.DataFrame(result)
```

```python
df_test.rename(columns={0:'content'}, inplace = True)
df_test
```
