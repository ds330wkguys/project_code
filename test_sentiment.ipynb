```python
# import dependencies
import pandas as pd
import snscrape.modules.twitter as sntwitter
import itertools
from nltk import *
import re
import string
from textblob import *
from wordcloud import *
import preprocessor as p
import os
```

```python
def Twitter_Data(search_word, start_day, end_day, nums_tweets):
    
    def getSubjectivity(text):
        return TextBlob(text).sentiment.subjectivity

    # polarity: the actual score
    def getPolarity(text):
        return TextBlob(text).sentiment.polarity

    def getAnalysis(score):
        if score < 0:
            return 'Negative'
        elif score == 0:
            return 'Neutral'
        else:
            return 'Positive'
    
    
    search_word = str(search_word)
    start_day = str(start_day)
    end_day = str(end_day)
    nums_tweets = int(nums_tweets)
    
    search_query = search_word + ' since:' + start_day + ' until:' + end_day 
    
    scraped_tweets = sntwitter.TwitterSearchScraper(search_query).get_items()
    sliced_scraped_tweets = itertools.islice(scraped_tweets, nums_tweets)
    
    df = pd.DataFrame(sliced_scraped_tweets)
    index = [i for i in range(len(df))]
    index_df = pd.DataFrame(index)
    df['index'] = index_df
    
    df_tmp = df[['index','date','renderedContent']].copy()
    list_content = []
    for i in df_tmp['renderedContent']:
        list_content.append(i)
        
    tmp = []
    for i in list_content:
        tmp.append(p.clean(i))
        
    clean_df = pd.DataFrame(tmp)
    df_tmp['renderedContent'] = clean_df[0]
    
    result = []
    for i in tmp:
        text = re.sub('[^a-zA-Z0-9]',' ',i).strip()
        text = re.sub(r'[^\w\s]', '', text)
        text = re.sub(r'#','',text)
        text = re.sub(r'RT[\s]+','',text)
        text = re.sub(r'https?:\/\/\S+','',text)
        if(text != ''):
            result.append(text)
        
    df_result = pd.DataFrame(result)
    
    index1 = [j for j in range(len(df_result))]
    index2 = pd.DataFrame(index1)
    df_result['index'] = index2
    
    df_left = pd.merge(df_tmp, df_result, left_on = 'index', right_on='index', how='left')
    df_left = df_left.drop('renderedContent', axis = 'columns').dropna()
    
    df_left = df_left.drop('index', axis = 'columns')
    df_left['date'] = pd.to_datetime(df_left['date'])
    df_left['date'] = df_left['date'].dt.strftime('%Y-%m-%d')
    df_left.rename(columns = {0:'text'},inplace=True)
    
    df_left['subjectivity'] = df_left['text'].apply(getSubjectivity)
    df_left['Polarity'] = df_left['text'].apply(getPolarity)
    
    df_left['score'] = df_left['Polarity'].apply(getAnalysis)

    PATH = os.getcwd()
    
    df_left.to_csv(PATH + '/' + search_word + '.csv')
    
    print('Success!')
```
